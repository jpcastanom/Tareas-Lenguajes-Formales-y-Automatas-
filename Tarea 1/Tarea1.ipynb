**Fig. 3.19.** The objective is that $b_1 b_2 \ldots b_{f(s)}$ is the longest proper prefix of $b_1 b_2 \ldots b_s$ that is also a suffix of $b_1 b_2 \ldots b_s$. The reason $f(s)$ is important is that if we are trying to match a text string for $b_1 b_2 \ldots b_n$, and we have matched the first $s$ positions but we then fail (i.e., the next position of the text string does not hold $b_{s+1}$), then $f(s)$ is the longest prefix of $b_1 b_2 \ldots b_n$ that could possibly match the text string up to the point we are at. Of course, the next character of the text string must be $b_{f(s)+1}$, or else we still have problems and must consider a yet shorter prefix, which will be $b_{f(f(s))}$.

---

### Algorithm

$$
\begin{aligned}
1)\ & t = 0 \\
2)\ & f(1) = 0 \\
3)\ & \text{for } (s = 1; s < n; s++) \{ \\
4)\ & \quad \text{while } (t > 0 \ \&\&\ b_{s+1} \ne b_{t+1})\ t = f(t); \\
5)\ & \quad \text{if } (b_{s+1} = b_{t+1}) \{ \\
6)\ & \qquad t = t + 1; \\
7)\ & \qquad f(s+1) = t; \\
8)\ & \quad \} \ \text{else } f(s+1) = 0; \\
9)\ & \}
\end{aligned}
$$

Algorithm to compute the failure function for keyword $b_1 b_2 \ldots b_n$

---

### Example

As an example, the failure function for the trie constructed above for **ababaa** is:

$$
\begin{array}{c|cccccc}
s & 1 & 2 & 3 & 4 & 5 & 6 \\
\hline
f(s) & 0 & 0 & 1 & 2 & 3 & 1
\end{array}
$$

For instance, states $3$ and $1$ represent prefixes **aba** and **a**, respectively.  
$f(3) = 1$ because **a** is the longest proper prefix of **aba** that is also a suffix of **aba**.  
Also, $f(2) = 0$, because the longest proper prefix of **ab** that is also a suffix is the empty string.

### Exercise 3.4.3

Construct the failure function for the strings:

- a) **abababaab**  
- b) **aaaaaa**  
- c) **abbaabb**

```python
def compute_failure_function(P):

    """Args:
        P: cadena de texto (el patrón).

    Returns:
        Una lista que representa la función de falla, donde f[i] es la longitud del
        prefijo propio más largo de P[0...i] que también es sufijo de P[0...i].
        y f[0] corresponde a f(1) en la notación base-1.
    """
    m = len(P)

    if m == 0:
        return []

    f = [0] * m
    t = 0 # registro de la longitud del prefijo y sufijo

    f[0] = 0

    if m == 1:
        return f

    # Calcula f[i] para i desde 1 hasta m-1
    for i in range(1, m):
        # Desplazamiento si no hay coincidencia
        while t > 0 and P[i] != P[t]:
            t = f[t-1]

        # Coincidencia, extiende el prefijo/sufijo
        if P[i] == P[t]:
            t = t + 1
            f[i] = t
        else:
            f[i] = 0 # No hay coincidencia y t ya es 0

    return f

def print_failure_table(pattern, failure_function):
    """Imprime la tabla de la función de falla formateada de  forma sencilla."""
    print(f"Patrón: '{pattern}'")
    print(f"Función de Falla: {failure_function}")


print("Ejercicios 3.4.3:")

# Ejercicio 3.4.3 a) abababaab
pattern_a = "abababaab"
f_a = compute_failure_function(pattern_a)
print(f"a) '{pattern_a}'")
print_failure_table(pattern_a, f_a)
print("\n")

# Ejercicio 3.4.3 b) aaaaaa
pattern_b = "aaaaaa"
f_b = compute_failure_function(pattern_b)
print(f"b) '{pattern_b}'")
print_failure_table(pattern_b, f_b)
print("\n")

# Ejercicio 3.4.3 c) abbaabb
pattern_c = "abbaabb"
f_c = compute_failure_function(pattern_c)
print(f"c) '{pattern_c}'")
print_failure_table(pattern_c, f_c)
```
