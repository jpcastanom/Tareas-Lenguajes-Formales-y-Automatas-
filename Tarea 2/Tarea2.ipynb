Having computed the failure function for a keyword $b_1 b_2 \cdots b_n$, we can scan a string $a_1 a_2 \cdots a_m$ in time $O(m)$ to tell whether the keyword occurs in the string. The algorithm slides the keyword along the string, trying to make progress by matching the next character of the keyword with the next character of the string. If it cannot do so after matching $s$ characters, then it “slides” the keyword right $s - f(s)$ positions, so only the first $f(s)$ characters of the keyword are considered matched with the string.

$$
\begin{aligned}
1)\;& s = 0; \\
2)\;& \textbf{for } (i = 1;\ i \leq m;\ i++)\ \{ \\
3)\;& \qquad \textbf{while } (s > 0 \ \&\&\ a_i \neq b_{s+1})\ s = f(s); \\
4)\;& \qquad \textbf{if } (a_i = b_{s+1})\ s = s + 1; \\
5)\;& \qquad \textbf{if } (s = n)\ \textbf{return } \text{“yes”}; \\
\;& \} \\
6)\;& \textbf{return } \text{“no”};
\end{aligned}
$$

The KMP algorithm tests whether string $a_1 a_2 \cdots a_m$ contains a single keyword $b_1 b_2 \cdots b_n$ as a substring in $O(m+n)$ time.

---

**Exercise 3.4.6:** Apply Algorithm KMP to test whether keyword $ababaa$ is a substring of:

a) $abababaab$.

b) $abababbaa$.


```python
def compute_failure(keyword: str) -> list[int]:
    """
    Compute the KMP failure function for the given keyword.

    Args:
        keyword: The pattern string b1b2...bn.

    Returns:
        A list f where f[s] is the length of the longest proper prefix of
        keyword[:s+1] that is also a suffix of keyword[:s+1].
        f[0] is always 0 by definition (no proper prefix for a single character).
    """
    n = len(keyword)
    f = [0] * n  # f[0] = 0 always

    k = 0  # length of the previous longest prefix-suffix
    for i in range(1, n):
        while k > 0 and keyword[k] != keyword[i]:
            k = f[k - 1]
        if keyword[k] == keyword[i]:
            k += 1
        f[i] = k

    return f

def kmp_search(text: str, keyword: str) -> bool:
    """
    Test whether `keyword` is a substring of `text` using the KMP algorithm
    (Figure 3.20).

    Args:
        text:    The string to search in.
        keyword: The pattern to search for.

    Returns:
        True if keyword appears as a substring of text, False otherwise.
    """
    if not keyword:
        return True
    if not text:
        return False

    m = len(text)
    n = len(keyword)
    f = compute_failure(keyword)

    s = 0  # number of characters currently matched

    for i in range(m):
        while s > 0 and text[i] != keyword[s]:
            s = f[s - 1]
        if text[i] == keyword[s]:
            s += 1
        if s == n:
            return True

    return False

print("Exercice 3.4.6 solution")
print(f"\n{'─' * 60}")

keyword = "ababaa"
test_strings = [("a", "abababaab"),("b", "abababbaa")]
# Show failure function
f = compute_failure(keyword)
print(f"\nKeyword : {keyword}")
print(f"Indices : {list(range(len(keyword)))}")
print(f"Failure : {f}")

for label, text in test_strings:
        result = kmp_search(text, keyword)
        answer = "YES — keyword is a substring" if result else "NO  — keyword is NOT a substring"
        print(f"\n{'─' * 60}")
        print(f"Part {label}) text = \"{text}\"")
        print(f"Result  : {answer}")
```
